```python
import sys
import traceback

def get_error_info():
    """Helper function to extract current exception details."""
    _, _, ex_tb = sys.exc_info()
    # Extract the last frame of the traceback
    summary = traceback.extract_tb(ex_tb)[-1]
    return summary.name, summary.lineno

def divide(a, b):
    try:
        result = a / b
        return result
    except Exception as e:
        # Call our helper to get the details
        func_name, line_no = get_error_info()
        
        print(f"--- Error Detected ---")
        print(f"Message: {e}")
        print(f"Function: {func_name}()")
        print(f"Line Number: {line_no}")
        return None
    
# Test the function
divide(10, 0)
```

```python
import pandas as pd

data = {
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT', 'IT'],
    'Salary': [50000, 60000, 45000, 55000, 70000, 80000],
    'Bonus': [2000, 3000, 1000, 1500, 5000, 6000]
}

df = pd.DataFrame(data)
print(df)
```

```python
## Group the data by 'Department' and calculate the TOTAL Salary (cost to company) and the AVERAGE Bonus (performance metric) 
# for each department.
```

```python
result= df.groupby("Department").agg(Total_Salary=("Salary", "sum"),Average_Bonus=("Bonus","mean"))
print(result)
```
