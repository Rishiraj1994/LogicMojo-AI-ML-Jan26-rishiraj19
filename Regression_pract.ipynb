```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Scikit-learn datasets and tools
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
data= load_diabetes(as_frame=True)
```

```python
df=data.frame
```

```python
df[]
```

```python
X_df = data.data.copy()
y = data.target.copy()
```

```python
df = X_df.copy()
df["target"] = y
```

```python
df.head()
```

```python
y
```

```python
X_df
```

```python
x_train, x_test, y_train,y_test = train_test_split(X_df,y, test_size=0.2 ,random_state=42)
```

```python
mean_guess= y_train.mean()
mean_guess
```

```python
baseline_predictios= np.full(shape=y_test.shape,fill_value=mean_guess)
```

```python
baseline_predictios
```

```python
baseline_mae= mean_absolute_error(y_test,baseline_predictios )
baseline_mae
```

```python
lin_reg= LinearRegression()
```

```python
lin_reg.fit(x_train,y_train)
```

```python
y_pred= lin_reg.predict(x_test)
```

```python
mae= mean_absolute_error(y_test, y_pred)
rsme= mean_squared_error(y_test,y_pred)
r2score= r2_score(y_test,y_pred)
```

```python
mae,rsme,r2score
```

```python
alphas= np.logspace(-4,4,100)
```

```python
alphas
```

```python
ridge_model = Pipeline([
("scaler",StandardScaler()),
("ridge",RidgeCV(alphas=alphas, cv=5))


])

lasso_model = Pipeline([
    ("scaler", StandardScaler()),
    ("lasso", LassoCV(alphas=alphas, cv=5, max_iter=20000))
])
```

```python
ridge_model.fit(x_train, y_train)
lasso_model.fit(x_train, y_train)
```

```python
lasso_weights = lasso_model.named_steps["lasso"].coef_
zero_weights_count = sum(abs(lasso_weights) < 1e-7)
```
