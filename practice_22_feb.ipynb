```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y1 = [10,20,15,25]
y2 = [5,15,10,20]

plt.figure(figsize=(8,5))

# First subplot
plt.subplot(1,2,1)
plt.plot(x, y1)
plt.title("First Plot")

# Second subplot
plt.subplot(1,2,2)
plt.plot(x, y2)
plt.title("Second Plot")

plt.show()
```
